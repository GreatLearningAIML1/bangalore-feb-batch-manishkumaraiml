2. Use the five percent significance test over the data to determine the p value to check new scheme has significantly raised outputs?

```python
import pandas as pd
from scipy.stats import ttest_ind,ttest_1samp, wilcoxon,stats
old_scheme = data["Old Scheme (in thousands)"]
new_scheme = data["New Scheme (in thousands)"]

#t_statistic, p_value = ttest_ind(old_scheme, new_scheme)
t_statistic, p_value =stats.ttest_rel(old_scheme,new_scheme)

print(t_statistic, p_value)
print("p_value ",p_value)
print("Accepting H0")
print("there is no new scheme has significantly raised outputs")
```
